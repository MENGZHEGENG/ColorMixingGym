```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from plotly.colors import DEFAULT_PLOTLY_COLORS as colors
import numpy as np
import json
from dist import state_similarity
```

```python
def make_old_plot(df, title):
    fig = make_subplots(rows=4, cols=1, shared_xaxes=True, vertical_spacing=0.02)

# Calculate breakpoints for the sections
    total_levels = len(df['filename'])

# Adding line plots instead of scatter plots
    fig.add_trace(
        go.Scatter(x=df['filename'], y=df['num_attempts'], mode='lines+markers', 
                #    line=dict(color='blue'), # You can customize the line color or style
                #    marker=dict(symbol=df['pass_fail'].map({True: 'circle', False: 'x'}),
                #                size=5),
                   marker=dict(color=df['passed'].map({True: 'green', False: 'red'})),

                #    marker_symbol=,

                   line=dict(color=colors[0]), # Customize color
                   showlegend=False), # Hides the legend for this trace
        row=1, col=1
    )

    fig.add_trace(
        go.Scatter(x=df['filename'], y=df['num_llm_queries'], mode='lines+markers', 
                   line=dict(color=colors[0]), # Customize color
                   marker=dict(color=df['passed'].map({True: 'green', False: 'red'})),
                   showlegend=False), # Hide legend
        row=2, col=1
    )

    fig.add_trace(
        go.Scatter(x=df['filename'], y=df['experience_sim_score'], mode='lines+markers', 
                #    line=dict(color='green'), # Customize color
                   line=dict(color=colors[0]), # Customize color
                   marker=dict(color=df['passed'].map({True: 'green', False: 'red'})),
                   showlegend=False), # Hide legend
        row=3, col=1
    )

    fig.add_trace(
        go.Scatter(x=df['filename'], y=df['fixed_tc_sim_score'], mode='lines+markers', 
                #    line=dict(color='green'), # Customize color
                   line=dict(color=colors[0]), # Customize color
                   marker=dict(color=df['passed'].map({True: 'green', False: 'red'})),
                   showlegend=False), # Hide legend
        row=4, col=1
    )

    goals = []
    for x in list(df["filename"]):
        f = open(f"data/level1/{x}")
        yes = json.load(f)
        goals.append(yes)
    
    last_states = []    
    for x in list(df["last_trajectory"]):
        a = eval(x)
        last_states.append(a[-1]["actual_next_state"])
    
    scores = []
    for goal, last_state in zip(goals, last_states):
        scores.append(state_similarity(goal, last_state))
        
    fig.add_trace(
        go.Scatter(x=df['filename'], y=scores, mode='lines+markers', 
                #    line=dict(color='green'), # Customize color
                   line=dict(color=colors[0]), # Customize color
                   marker=dict(color=df['passed'].map({True: 'green', False: 'red'})),
                   showlegend=False), # Hide legend
        row=4, col=1
    )
    

    

# Update layout to remove legend and adjust titles
    fig.update_layout(height=800, width=800, title_text=title, showlegend=False)
    fig.update_yaxes(title_text="Num Attempts", row=1, col=1)
    fig.update_yaxes(title_text="Num LLM Queries", row=2, col=1)
    fig.update_yaxes(title_text="Experience Sim Score", row=3, col=1, range=[-0.1,1.1])
    fig.update_yaxes(title_text="Fixed data set Sim Score", row=4, col=1, range=[-0.1,1.1])
    fig.update_xaxes(title_text="Level", row=4, col=1)

# Show the figure
    fig.show(renderer='browser')
```

```python
fig = make_subplots(rows=4, cols=2, shared_xaxes=True, vertical_spacing=0.02)

# Colors for different sections
first_section_color = 'rgba(135, 206, 235, 0.2)'  # Light blue with some transparency
middle_section_color = 'rgba(144, 238, 144, 0.2)'  # Light green with some transparency
last_section_color = 'rgba(240, 128, 128, 0.2)'  # Light coral with some transparency


# Calculate breakpoints for the sections
total_levels = len(df['filename'])
section_size = total_levels // 3  # Assuming 30 levels as per the example data


# Helper function to add a tinted background for a section
def add_section_background(fig, start_level, end_level, color, rows=[1, 2, 3]):
   for row in rows:
       fig.add_shape(
           type="rect",
           xref="x" + str(row), yref="paper",
           x0=df['filename'].iloc[start_level], x1=df['filename'].iloc[end_level],
           y0=0, y1=1,
           fillcolor=color,
           opacity=0.5,
           layer="below",
           line_width=0,
       )


# Add background tinting
add_section_background(fig, 0, section_size-1, first_section_color)
add_section_background(fig, section_size, section_size*2-1, middle_section_color)
add_section_background(fig, section_size*2, total_levels-1, last_section_color)


# Adding line plots instead of scatter plots
fig.add_trace(
   go.Scatter(x=df['filename'], y=df['num_attempts'], mode='lines+markers',
           #    line=dict(color='blue'), # You can customize the line color or style
           #    marker=dict(symbol=df['pass_fail'].map({True: 'circle', False: 'x'}),
           #                size=5),
              marker=dict(color=df['passed'].map({True: 'green', False: 'red'})),
              line=dict(color=colors[0]), # Customize color
              showlegend=False), # Hides the legend for this trace
   row=1, col=1
)


fig.add_trace(
   go.Scatter(x=df['filename'], y=df['num_llm_queries'], mode='lines+markers',
              line=dict(color=colors[0]), # Customize color
              marker=dict(color=df['passed'].map({True: 'green', False: 'red'})),
              showlegend=False), # Hide legend
   row=2, col=1
)


fig.add_trace(
   go.Scatter(x=df['filename'], y=df['experience_PR'], mode='lines+markers',
           #    line=dict(color='green'), # Customize color
              line=dict(color=colors[0]), # Customize color
              marker=dict(color=df['passed'].map({True: 'green', False: 'red'})),
              showlegend=False), # Hide legend
   row=3, col=1
)


categories = ['stack_PR', 'unstack_PR', 'pick_up_PR', 'put_down_PR']
df['average_pr'] = df[categories].mean(axis=1)


fig.add_trace(
   go.Scatter(x=df['filename'], y=df['average_pr'], mode='lines+markers',
           #    line=dict(color='green'), # Customize color
              line=dict(color=colors[0]), # Customize color
              marker=dict(color=df['passed'].map({True: 'green', False: 'red'})),
              showlegend=False), # Hide legend
   row=4, col=1
)

fig.add_trace(
   go.Scatter(x=df['filename'], y=df['stack_PR'], mode='lines+markers',
           #    line=dict(color='green'), # Customize color
              line=dict(color=colors[0]), # Customize color
              marker=dict(color=df['passed'].map({True: 'green', False: 'red'})),
              showlegend=False), # Hide legend
   row=1, col=2
)


fig.add_trace(
   go.Scatter(x=df['filename'], y=df['unstack_PR'], mode='lines+markers',
           #    line=dict(color='green'), # Customize color
              line=dict(color=colors[0]), # Customize color
              marker=dict(color=df['passed'].map({True: 'green', False: 'red'})),
              showlegend=False), # Hide legend
   row=2, col=2
)


fig.add_trace(
   go.Scatter(x=df['filename'], y=df['pick_up_PR'], mode='lines+markers',
           #    line=dict(color='green'), # Customize color
              line=dict(color=colors[0]), # Customize color
              marker=dict(color=df['passed'].map({True: 'green', False: 'red'})),
              showlegend=False), # Hide legend
   row=3, col=2
)


fig.add_trace(
   go.Scatter(x=df['filename'], y=df['put_down_PR'], mode='lines+markers',
           #    line=dict(color='green'), # Customize color
              line=dict(color=colors[0]), # Customize color
              marker=dict(color=df['passed'].map({True: 'green', False: 'red'})),
              showlegend=False), # Hide legend
   row=4, col=2
)


# Update layout to remove legend and adjust titles
fig.update_layout(height=800, width=1600, title_text="GPT3 Results", showlegend=False)
fig.update_xaxes(title_text="Level", row=4, col=1)
fig.update_xaxes(title_text="Level", row=4, col=2)
fig.update_yaxes(title_text="Num Attempts", row=1, col=1)
fig.update_yaxes(title_text="Num LLM Queries", row=2, col=1)
fig.update_yaxes(title_text="Experience_pr", row=3, col=1, range=[-0.1,1.1])
fig.update_yaxes(title_text="Frozen_pr", row=4, col=1, range=[-0.1,1.1])
fig.update_yaxes(title_text="stack_pr", row=1, col=2, range=[-0.1,1.1])
fig.update_yaxes(title_text="unstack_pr", row=2, col=2, range=[-0.1,1.1])
fig.update_yaxes(title_text="pickup_pr", row=3, col=2, range=[-0.1,1.1])
fig.update_yaxes(title_text="putdown_pr", row=4, col=2, range=[-0.1,1.1])


# Show the figure
fig.show(renderer='browser')
```

```python
# Calculate the Total Pass Rate as the average of the four sub pass rates

categories = ['stack_PR', 'unstack_PR', 'pick_up_PR', 'put_down_PR']
df['Total_Pass_Rate'] = df[categories].mean(axis=1)

# Create the figure
fig = go.Figure()

# Add traces for each pass rate
colors = ['blue', 'green', 'red', 'purple']  # Different colors for each category

for category, color in zip(categories, colors):
    fig.add_trace(
        go.Bar(
            x=df.index,  # Assuming the index of the DataFrame should be used as x
            y=df[category] / 4 * 100,
            name=category.replace('_', ' ').title(),  # Cleaning up the name for the legend
            marker_color=color
        )
    )

# Change the bar mode to stack to create a stacked bar plot
fig.update_layout(
    barmode='stack',
    title='Total Pass Rate with Contributions from Each Category',
    xaxis_title='Index',
    yaxis_title='Pass Rate',
    yaxis=dict(range=[0, 100])  # Ensure the y-axis ranges from 0 to 1
)

# Show the figure
fig.show(renderer='browser')
```
